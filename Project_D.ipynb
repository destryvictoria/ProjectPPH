## Data wrangling

### Subtask:
Data Wrangling and Feature Engineering


**Reasoning**:
Aggregate the data by year to facilitate year-over-year growth calculations and ratio analysis. Select a relevant subset of columns for aggregation that represent key financial metrics.



```python
df_yearly = df.groupby('tahun')[['pendapatan_operasional', 'beban_operasional', 'laba_bersih_normal', 'persediaan_awal', 'persediaan_akhir', 'total_hpp']].sum().reset_index()
display(df_yearly)
```

**Reasoning**:
Calculate year-over-year growth rates for key financial metrics and create new columns to store these growth rates. Handle the first year's growth rate appropriately.



```python
# Calculate year-over-year growth rates
for col in ['pendapatan_operasional', 'beban_operasional', 'laba_bersih_normal', 'total_hpp']:
    df_yearly[f'{col}_growth'] = df_yearly[col].pct_change() * 100
    df_yearly[f'{col}_growth'].fillna(0, inplace=True)  # Handle the first year

# Calculate financial ratios
df_yearly['profit_margin'] = (df_yearly['laba_bersih_normal'] / df_yearly['pendapatan_operasional']) * 100
df_yearly['operating_expense_ratio'] = (df_yearly['beban_operasional'] / df_yearly['pendapatan_operasional']) * 100
df_yearly['inventory_turnover'] = df_yearly['total_hpp'] / ((df_yearly['persediaan_awal'] + df_yearly['persediaan_akhir']) / 2)

display(df_yearly)
```

## Data visualization

### Subtask:
Visualize key financial trends and ratios from the yearly aggregated data.


**Reasoning**:
Visualize the key financial trends and ratios using line and bar charts, as well as a scatter plot to explore relationships between variables.



```python
import matplotlib.pyplot as plt

# Line chart for key financial metrics
plt.figure(figsize=(12, 6))
plt.plot(df_yearly['tahun'], df_yearly['pendapatan_operasional'], label='Pendapatan Operasional', color='blue')
plt.plot(df_yearly['tahun'], df_yearly['beban_operasional'], label='Beban Operasional', color='green')
plt.plot(df_yearly['tahun'], df_yearly['laba_bersih_normal'], label='Laba Bersih Normal', color='red')
plt.title('Tren Pendapatan Operasional, Beban Operasional, dan Laba Bersih Normal')
plt.xlabel('Tahun')
plt.ylabel('Nilai (IDR)')
plt.legend()
plt.grid(True)
plt.show()

# Bar chart for year-over-year growth rates
plt.figure(figsize=(12, 6))
X_axis = np.arange(len(df_yearly['tahun']))
plt.bar(X_axis - 0.2, df_yearly['pendapatan_operasional_growth'], 0.2, label='Pendapatan Operasional', color='blue')
plt.bar(X_axis, df_yearly['beban_operasional_growth'], 0.2, label='Beban Operasional', color='green')
plt.bar(X_axis + 0.2, df_yearly['laba_bersih_normal_growth'], 0.2, label='Laba Bersih Normal', color='red')
plt.xticks(X_axis, df_yearly['tahun'])
plt.title('Pertumbuhan Tahun ke Tahun')
plt.xlabel('Tahun')
plt.ylabel('Tingkat Pertumbuhan (%)')
plt.legend()
plt.grid(True)
plt.show()

# Line chart for financial ratios
plt.figure(figsize=(12, 6))
plt.plot(df_yearly['tahun'], df_yearly['profit_margin'], label='Profit Margin', color='blue')
plt.plot(df_yearly['tahun'], df_yearly['operating_expense_ratio'], label='Operating Expense Ratio', color='green')
plt.plot(df_yearly['tahun'], df_yearly['inventory_turnover'], label='Inventory Turnover', color='red')
plt.title('Tren Rasio Keuangan')
plt.xlabel('Tahun')
plt.ylabel('Rasio')
plt.legend()
plt.grid(True)
plt.show()


# Scatter plot to explore relationship between revenue and profit
plt.figure(figsize=(8, 6))
plt.scatter(df_yearly['pendapatan_operasional'], df_yearly['laba_bersih_normal'], c=df_yearly['tahun'], cmap='viridis')
plt.title('Hubungan Antara Pendapatan Operasional dan Laba Bersih')
plt.xlabel('Pendapatan Operasional')
plt.ylabel('Laba Bersih')
plt.colorbar(label='Tahun')
plt.grid(True)
plt.show()
```
